![alt text](https://i.imgur.com/Xc6CrOy.jpg "Logo Title Text 1")

- Startup can send a model spec to a smart contract
- The smart contract intializes the model itself. Because the wieghts are randomly intialized by the smart contract and all the information used to intilaize the weights is public, there are no hidden secrets inside the homomorphic encryption that our startup could embed to make it so that the model secretly memorizes the data. 
- The smart contract sends the model spec to the oracle
- The oracle generates a public private key pair, encrypts the model, then sends it back to the smart contract
- After training on all the users, the model is sent to the oracle, its decrypted, encrypted with the oracle’s keys, then sent back. 
- Since the oracle acts as a trusted third party, no one can decrypt the model or gradients. 
- A second smart contract that acts as an insurance policy against the oracle sharing secrets.  

- 启动可以向智能合同发送模型规范

# Proof of work 工作证明

![alt text](https://image.slidesharecdn.com/blockchainbtsym-161201162319/95/blockchain-btsym-16-11-638.jpg?cb=1480610107 "Logo Title Text 1")

![alt text](https://image.slidesharecdn.com/blockchainbtsym-161201162319/95/blockchain-btsym-16-12-638.jpg?cb=1480610107 "Logo Title Text 1")

```python
#Simple Proof-of-Work example

import sys
import time
import hashlib 
from struct import unpack, pack

#The timestamp, message and payload are just the “stuff” that you 
#want to send the network. 
#It could be block data, as with Bitcoin mining.

timestamp = str(time.time())
message = "this is a random message"
payload = timestamp + message

#The nonce, guess, throttle and target are used to perform the work. 
#The important thing about proof-of-work is that it has to be difficult 
#for the 
#client but easy for the system to check. Only the nonce changes with 
#each guess. 
#The probability of being correct is so low that this system, 
#probabilistically, 
#guarantees many guesses or work, is performed.

nonce = 0
guess = 999999999999
throttle = 100000000
target = 2**64/ throttle

#
#nonce increments by 1 each time. 
#Effectively counting how many guesses the computer makes. 
#However, that is not a requirement.

#When the system checks the work, 
#it does not have to trace back over each guess. 
#It only checks the nonce submitted with the data.


payloadHash = hashlib.sha512(payload).digest()

start = time.time()

#These three lines are our proof-of-work algorithm. 
#It’s a simple loop. This hashes our data with two rounds of SHA256. 
#The first 8 bytes are turned in to a number. 
#Effectively, this is how a computer makes a guess.
while guess > target:
    nonce+1
    guess, = unpack('>Q', hashlib.sha512(hashlib.sha512(pack('>Q', nonce) +
                                                       payloadHash).digest()).digest())[0:8]
end = time.time()
#The nonce is what proves the work was done. 
#In this example, it’s value represents the number of attempts your CPU made before 
#it found a valid guess below the target. Because each guess has the same probability of being 
#under the target the method of generating your nonce does not matter. However, simply incrementing 
#the nonce is cheaper than generating a random number. When work is submitted to the network, 
#the nonce is used to verify the correctness.
print "%s:%s:%s:%s:%s:%s:%s" % (timestamp, message, nonce, guess, payload, target, end-start)
```

# Proof of Stake 股权证明

![alt text](http://webcomsystem.net/blog/wp-content/uploads/2017/08/POW-vs-POS.jpg "Logo Title Text 1")

- Proof of Stake (PoS) is a category of consensus algorithms for public blockchains that depend on a validator's economic stake in the network. 
- In proof of work (PoW) based public blockchains (e.g. Bitcoin and the current implementation of Ethereum), the algorithm rewards participants who solve cryptographic puzzles in order to validate transactions and create new blocks (i.e. mining). 
- In PoS-based public blockchains (e.g. Ethereum's upcoming Casper implementation), a set of validators take turns proposing and voting on the next block, and the weight of each validator's vote depends on the size of its deposit (i.e. stake). 
- Significant advantages of PoS include security, reduced risk of centralization, and energy efficiency.

- 股权证明(PoS)是公共区块链的一种共识算法, 它依赖于验证器在网络中的经济利益。
- 在工作证明(PoW)为基础的公共区块链(如比特币和目前的实施以太坊), 算法奖励参与者谁解决密码难题, 以验证事务和创建新的区块 (即采矿)。
- 在基于PoS公共区块链(如以太坊即将推出的小精灵实施) 中, 一组验证器轮流提议并在下一个区块投票, 每个验证程序的投票权取决于其存款的大小 (即股权)。
- PoS的显著优势包括安全性、降低集中化风险和能效。

In general, a proof of stake algorithm looks as follows. 

一般情况下, 股权证明算法 如下。

- The blockchain keeps track of a set of validators
- Anyone can become a validator by sending a special type of transaction that locks up their coin into a deposit
- The process of creating and agreeing to new blocks is then done through a consensus algorithm that all current validators can participate in
- Many ways to assign rewards to validators who participate in the consensus algorithm, two major types: chain-based proof of stake and BFT-style proof of stake.

- 区块链跟踪一组验证器
- 任何人都可以成为验证器，只要通过发送一种特殊类型的交易将他们的硬币锁定在存款中来
- 创建和同意新区块的过程, 然后通过一个一致性算法, 所有当前的验证程序可以参与
- 有许多方法可以将奖励分配给参与协商一致算法的验证者, 两种主要类型: 链式的股权证明 和 BFT形式的股权证明。

In chain-based proof of stake

在链证明的股权

- the algorithm pseudo-randomly selects a validator during each time slot (eg. every period of 10 seconds might be a time slot)
- assigns that validator the right to create a single block
- this block must point to some previous block (normally the block at the end of the previously longest chain), and so over time most blocks converge into a single constantly growing chain.

- 算法pseudo-randomly在每个时间槽中选择一个验证器 (例如,每10秒可能是一个时间段)
- 赋予验证程序创建单个区块的权限
- 此区块必须指向以前的区块 (通常是以前最长的链的末尾的区块), 因此随着时间的推移, 大多数区块会聚成一个不断增长的单一链。

In BFT-style proof of stake (byzantine fault tolerant)

在BFT样式证明的股权 (拜占庭式的容错)

- validators are randomly assigned the right to propose blocks
- agreeing on which block is canonical is done through a multi-round process where every validator sends a "vote" for some specific block during each round
- at the end of the process all (honest and online) validators permanently agree on whether or not any given block is part of the chain. 
- Blocks may still be chained together; the key difference is that consensus on a block can come within one block, and does not depend on the length or size of the chain after it.

- 验证程序随机分配了建议区块的权限
- 同意在哪个区块是规范的是通过一个多的过程, 每个验证器发送一个 "投票" 的一些特定区块在每一个回合
- 在过程结束时, 所有 (诚实和在线) 验证程序永久同意是否任何给定的区块是链的一部分。
- 区块可能仍然被拴在一起;关键的区别在于, 区块上的一致意见可以在一个区块内出现, 而不取决于它后面的链的长度或大小。

活动证明 (混合PoW+ PoS)- DeCred governenance 币

- 区块不包含任何交易, 因此获奖区块将只包含一个标题和矿工的奖励地址
- 然后, 系统切换到股权证明

- 验证程序拥有的系统中的硬币越多, 就越有可能选择他或她。

### Proof of Burn 燃烧证明

- Instead of pouring money into expensive computer equipment, you 'burn' coins by sending them to an address where they are irretrievable. 
- By committing your coins to never-never land, you earn a lifetime privilege to mine on the system based on a random selection process.
- Over time, your stake in the system decays, so eventually you will want to burn more coins to increase your odds of being selected in the lottery. (This mimics bitcoin's mining process, where you have to continually invest in more modern computing equipment to maintain hashing power.)

- 代替把钱投入昂贵的电脑设备, 你"燃烧"硬币发送到一个不可挽回的地址。
- 通过承诺你的硬币永远不会着陆, 你获得了基于随机选择过程的系统中挖矿的终身的特权。
- 随着时间的推移, 你的股份在系统的衰变, 所以最终你会想燃烧更多的硬币, 来增加你被选中的几率。(这模仿比特币的采矿过程, 在那里你必须不断地投资于更多的现代计算设备来维持散列电力。)
